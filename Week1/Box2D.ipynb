```python
%matplotlib notebook 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
class Particle():
    
    # init
    def __init__(self, x0,v0,a0,t,m,radius,Id):
        
        self.dt  = t[1] - t[0]
        
        self.r = x0
        self.v = v0
        self.a = a0
        
        self.rVector = np.zeros( (len(t),len(x0)) )
        self.vVector = np.zeros( (len(t),len(v0)) )
        self.aVector = np.zeros( (len(t),len(a0)) )
        
        self.m = m
        self.radius = radius
        self.Id = Id
        
    # Method
    def Evolution(self):
        
        # Euler method
        self.r += self.dt * self.v
        self.v += self.dt * self.a
        
    def SetPosition(self,i,r):
        self.rVector[i] = r
        
    def SetVelocity(self,i,v):
        self.vVector[i] = v
```

```python
# Discretization
dt = 0.01
tmax = 10
t = np.arange(0,tmax+dt,dt)
print(len(t))
```

```python
def GetParticles(NParticles,Limit,Velo,Dim=2,dt=0.1):
    
    Particles_ = []
    
    for i in range(NParticles):
        
        x0 = np.random.uniform( -Limit+1.0, Limit-1.0, size=Dim )
        v0 = np.random.uniform( -Velo, Velo, size=Dim)
        a0 = np.zeros(Dim)
        
        p = Particle(x0,v0,a0,t,1.,1.0,i)
        
        Particles_.append(p)
        
    return Particles_
```

```python
Limits = np.array([20.,20.])
```

```python
Particles = GetParticles(100,Limits[0],2,dt=dt)
```

```python
for p in Particles:
    plt.scatter(p.r[0],p.r[1])
    plt.arrow(p.r[0],p.r[1],p.v[0],p.v[1],color='r',head_width=0.5)
```
